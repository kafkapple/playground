```python
import glob
def data_csv_read(path):
    csv_file = glob.glob(path+'*.csv')[0]
    #print(csv_file)
    time_bin = 9 # in case of OFL
    df = pd.read_csv(csv_file, header=None, skiprow=3).iloc[:,0:time_bin+1]
    df.rename(columns={0:'subject'}, inplace=True)
    
    key_subject = df.keys()[0] # pandas  csv read ,  key  ( row)

    obs = df[df[key_subject].str.contains("dem | exclude")==False]  # dem, onset, duration      -> obs   
    dem = df[df[key_subject].str.contains("dem")==True]  # dem data
    
    dem = dem.assign(group='dem')
    obs = obs.assign(group='obs')

    t = [obs,dem]
    pd.concat(t).sort_values(['subject', 'group']).reset_index(drop=True)

   
    return obs, dem

path_ofl = '/python/data/ofl/'
obs, dem = data_ofl_read(path_ofl)

```
